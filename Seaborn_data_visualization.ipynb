```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/titanic-training-data.csv')
df.info()
df.shape
df.head()
```

```python
#scatter plot
sns.relplot(x = "Age", y = "Fare", data = df)
```

```python
#catagory plot - hue plot
sns.catplot(x="Pclass", y="Fare", hue = "Sex", data=df)
```

```python
#catagory plot - jitter plot
sns.catplot(x="Pclass", y="Fare", data=df)
```

```python
#boxplot
sns.catplot(x="Pclass", y="Fare", kind = "box", hue = "Sex", data=df)
```

```python
#violinplot
sns.catplot(x="Pclass", y="Fare", hue = "Sex", kind = "violin", data=df)
sns.catplot(x="Pclass", y="Fare", hue = "Sex", kind = "violin", split = True, data=df)
```

```python
#barplot
sns.catplot(x="Pclass", y="Fare", hue = "Sex", kind = "bar", data=df)
```

```python
#pointplot
sns.catplot(x="Pclass", y="Fare", hue = "Sex", kind = "point", data=df)
```

```python
#swarmplot
sns.catplot(x="Pclass", y="Fare", hue="Sex", aspect=.9, kind="swarm", data=df);
```
